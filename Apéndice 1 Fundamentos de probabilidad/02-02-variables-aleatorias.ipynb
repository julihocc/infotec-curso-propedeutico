```python
# Variables aleatorias con NumPy
# Ejemplo 1: Distribución normal

import numpy as np
import matplotlib.pyplot as plt

# Generar 1000 valores aleatorios con distribución normal
mu = 100
sigma = 10
n = 1000
s = np.random.normal(mu, sigma, n)

# Histograma
count, bins, ignored = plt.hist(s, 30, density=True)
campana_gaussiana = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(- (bins - mu)**2 / (2 * sigma**2))
plt.plot(bins, campana_gaussiana,
         linewidth=2, color='r')
plt.show()
```

```python
# Ejemplo 2: Distribución binomial

n = 10
p = 0.5
s = np.random.binomial(n, p, 1000)

bins = np.arange(0, 12) - 0.5

# Histograma
count, bins, ignored = plt.hist(s, bins, density=True)
plt.show()
```

```python
# Ejemplo 3: Distribución Poisson

lam = 5
s = np.random.poisson(lam, 1000)

max_s = np.max(s)
print(max_s)

bins = np.arange(0, max_s+2) - 0.5

# Histograma
count, bins, ignored = plt.hist(s, bins, density=True)
plt.show()
```
