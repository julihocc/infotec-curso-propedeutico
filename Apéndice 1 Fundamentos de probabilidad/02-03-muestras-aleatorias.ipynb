```python
# Uso de Numpy para probabilidad y estadística	

import numpy as np

# Generar un arreglo de 1000 números aleatorios
data = np.random.normal(100, 10, 1000)

# Visualización de la simulación

import matplotlib.pyplot as plt

plt.hist(data, bins=50, color='lightblue', edgecolor='black')
plt.show()
```

```python
# Calcular la media 
mean = np.mean(data)
mean 
```

```python
# Calcular la mediana
median = np.median(data)
median
```

```python
# Calcular la varianza
variance = np.var(data)
variance
```

```python
# Calcular la desviación estándar
std_dev = np.std(data)
std_dev
```

```python
# Cuartiles
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1
iqr
```

```python
# Diagrama de caja
plt.boxplot(data)
plt.show()
```
