```python
# Aplicaciones Científicas de Numpy

## Simulaciones y modelos
### Uso de NumPy para simulaciones estocásticas y modelado matemático

# NumPy es una biblioteca muy útil para realizar simulaciones y modelado matemático.

# En este ejemplo, vamos a simular el lanzamiento de una moneda.

# Importamos NumPy
import numpy as np

# Definimos los posibles resultados
resultados = ['cara', 'sello']

# Simulamos el lanzamiento de una moneda
lanzamientos = np.random.choice(resultados)
print(lanzamientos)
```

```python
# Podemos simular el lanzamiento de una moneda un número determinado de veces
lanzamientos = np.random.choice(resultados, size=10)
print(lanzamientos)
```

```python
# Podemos simular el lanzamiento de una moneda con una probabilidad determinada
lanzamientos = np.random.choice(resultados, size=10, p=[0.2, 0.8])
print(lanzamientos)
```

```python
## Modelos matemáticos con distribuciones de probabilidad

# NumPy también nos permite trabajar con distribuciones de probabilidad.

# Por ejemplo, podemos generar números aleatorios con una distribución normal.
mu, sigma = 0, 0.1 # media y desviación estándar
s = np.random.normal(mu, sigma, 1000)

# Podemos visualizar la distribución de los números generados
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(s, 30, density=True)
gauss = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2))
plt.plot(bins, gauss, linewidth=2, color='r')
plt.show()
```

```python
# También podemos generar números aleatorios con una distribución uniforme
s = np.random.uniform(-1,0,1000)
count, bins, ignored = plt.hist(s, 15, density=True)
plt.plot(bins, np.ones_like(bins), linewidth=2, color='r')
plt.show()
```

```python
# O con una distribución exponencial
s = np.random.exponential(1, 1000)
count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, np.exp(-bins), linewidth=2, color='r')
plt.show()
```

```python
# Caso práctico con distribucion binomial

# Supongamos que lanzamos una moneda 10 veces y queremos saber cuántas veces sale cara.
# Podemos modelar este problema con una distribución binomial.
# La distribución binomial nos da la probabilidad de obtener k éxitos en n ensayos independientes.
# La probabilidad de obtener k éxitos en n ensayos es:
# P(X=k) = C(n, k) * p^k * (1-p)^(n-k)
# Donde C(n, k) es el número de combinaciones de n elementos tomados de k en k.
# En nuestro caso, n=10 y p=0.5.

# Simulamos el lanzamiento de una moneda 10 veces
lanzamientos = np.random.choice(resultados, size=10)
print(lanzamientos) 
```

```python
# Contamos cuántas veces sale cara
cara = np.count_nonzero(lanzamientos == 'cara')
print(cara)
```

```python
# Podemos repetir la simulación un número determinado de veces
n = 1000
cara = np.zeros(n)
for i in range(n):
    lanzamientos = np.random.choice(resultados, size=10)
    cara[i] = np.count_nonzero(lanzamientos == 'cara')  

# Visualizamos la distribución de los resultados
count, bins, ignored = plt.hist(cara, 10, density=True)
plt.show()
```

```python
# Calculamos la probabilidad de obtener 5 caras en 10 lanzamientos
# a partir de la simulación

# Contamos cuántas veces obtenemos 5 caras
cinco_caras = np.count_nonzero(cara == 5)
print(cinco_caras/n)
```

```python
# Podemos calcular la probabilidad de obtener k éxitos en n ensayos
from scipy.stats import binom
n = 10
p = 0.5
k = np.arange(0, 11)
prob = binom.pmf(k, n, p)
print(prob)
```

```python
# Extraemos la probabilidad de obtener 5 caras en 10 lanzamientos
print(prob[5])
```
