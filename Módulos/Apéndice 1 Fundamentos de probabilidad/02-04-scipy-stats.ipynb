```python
# Variables aleatorias con Scipy
# Ejemplo 1: Distribución normal

import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import norm

# Crear una variable aleatoria normal
data = norm.rvs(size=1000)

# Calcular la función de densidad de probabilidad
x = np.linspace(min(data), max(data), 100)
y = norm.pdf(x)

# Crear el histograma
plt.hist(data, bins=30, density=True, color='lightblue')
plt.plot(x, y, 'r--')
plt.show()
```

```python
# Ejemplo 2: Distribución binomial
from scipy.stats import binom

# Crear una variable aleatoria binomial
data = binom.rvs(n=10, p=0.5, size=1000)

# Calcular la función de densidad de probabilidad
x = np.arange(0, 11)
y = binom.pmf(x, n=10, p=0.5)

# Crear el histograma
bins = np.arange(0, 12) - 0.5
bins
```

```python

plt.hist(data, bins=bins, density=True, color='lightblue')
plt.scatter(x, y, color='red')
plt.show()
```

```python
# Ejemplo 3: Distribución de Poisson
from scipy.stats import poisson

# Crear una variable aleatoria de Poisson
data = poisson.rvs(mu=3, size=1000)


# Crear el histograma
min_data = np.min(data)
print(min_data)

max_data = np.max(data)
print(max_data)

# Calcular la función de densidad de probabilidad
x = np.arange(min(data), max(data)+1)
y = poisson.pmf(x, mu=3)

bins = np.arange(min(data), max(data)+2) - 0.5
```

```python

plt.hist(data, bins=bins, density=True, color='lightblue')
plt.scatter(x,y, color='red')
plt.show()
```
