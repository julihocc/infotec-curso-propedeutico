```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.normal(loc=10, scale=5, size=1000)
plt.hist(data)
plt.show()
```

```python
normalized_data = (data - data.mean()) / data.std()
np.mean(normalized_data)
```

```python
np.std(normalized_data)
```

```python
plt.hist(normalized_data)
plt.show()
```
