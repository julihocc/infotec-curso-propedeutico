# Eigenvectores y eigenvalores

```python
from sympy import Matrix

l1,l2,l3 = [2,3,3]
D = Matrix([[l1,0,0],[0,l2,0],[0,0,l3]])
D
```

```python
e1 = Matrix([1,0,0])
e2 = Matrix([0,1,0])
e3 = Matrix([0,0,1])

e1, e2, e3
```

```python
D*e1
```

```python
D*e2
```

```python
D*e3
```

```python
# De = ke <-> De-ke = 0

D*e1 - l1*e1
```

```python

D*e2 - l2*e2
```

```python

D*e3 - l3*e3
```

```python
P = Matrix([[1,1,1], [1,2,3], [1,3,6]])
P
```

```python
P.det()
```

```python
P.inv()
```

```python
A = P * D * P.inv()
A
```

```python
A.eigenvals()
```
