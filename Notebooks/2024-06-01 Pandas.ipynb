```python
relatives = df["SibSp"] + df["Parch"]
relatives
```

```python
target = df["Age"]
target = target.fillna(target.mean())

target
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(relatives.values.reshape(-1, 1), target.values)
y_pred = model.predict(relatives.values.reshape(-1,1))

plt.scatter(relatives, target)
plt.plot(relatives, y_pred, color="red")
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

target = target.astype(int) # convertimos a una variable categorica

model = LogisticRegression()
model.fit(relatives.values.reshape(-1, 1), target.values)
y_pred = model.predict(relatives.values.reshape(-1,1))

plt.scatter(relatives, target)
plt.scatter(relatives, y_pred, color="red")
plt.show()
```
