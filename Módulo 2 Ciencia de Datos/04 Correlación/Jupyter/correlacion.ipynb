```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
drift = 2
noise = np.random.normal(0, 1, size=len(x))
y = drift*x + noise

plt.scatter(x, y)
plt.show()
```

```python
correlation = np.cov(x, y) / (np.std(x) * np.std(y))

print(correlation)
```

```python
correlation = np.corrcoef(x, y)
print(correlation)
```

```python
x_norm = (x - x.mean()) / x.std()
y_norm = (y - y.mean()) / y.std()

correlation = np.cov(x_norm, y_norm)

print(correlation)
```
